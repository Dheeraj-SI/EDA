```python
#Average cost for 2 per Cuisine
Cuisine_avg_cost = df.groupby('Cuisines').mean()[['Average Cost for two']].reset_index()
Cuisine_avg_cost.sort_values(by='Average Cost for two',ascending=False,inplace=True)
Cuisine_avg_cost[0:10].plot(kind='barh',x='Cuisines',y='Average Cost for two')
```

```python
#Average Price Range in India for top 10 Cuisines
df_india = df[df['Country']=='India']
Cuisine_avg_cost = df_india.groupby('Cuisines').mean()[['Price range']].reset_index()
Cuisine_avg_cost.sort_values(by='Price range',ascending=False,inplace=True)
Cuisine_avg_cost[0:10].plot(kind='barh',x='Cuisines',y='Price range')
```

```python
#Cusines with highest number of votes
cusine_votes = df_india.groupby('Cuisines').sum()[['Votes']].reset_index()
cusine_votes.sort_values(by='Votes',ascending=False,inplace=True)
cusine_votes[0:10].plot(kind='barh',x='Cuisines',y='Votes',color='g',legend=False)
plt.title("Top ten Cusines with most number of Votes\n")
```

##### North Indian and Chinese cuisines have been most voted
